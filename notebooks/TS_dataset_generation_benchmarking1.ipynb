## Generate entire data set

```python
TS_classes = []
for filename in class_files:
    TS_classes.append(os.path.join(PATH_classes, filename))
```

```python
X_train, y_train = TSgen.generate_dataset(TS_classes,
                                      200,
                                      random_seed = None)

X_val, y_val = TSgen.generate_dataset(TS_classes,
                                      50,
                                      random_seed = None)

X_test, y_test = TSgen.generate_dataset(TS_classes,
                                      50,
                                      random_seed = None)
```

# Run mcfly

```python
DATA = 'C:\\OneDrive - Netherlands eScience Center\\Project_mcfly\\data\\processed'
CODE = 'C:\\OneDrive - Netherlands eScience Center\\Project_mcfly\\mcfly\\mcfly'
sys.path.append(CODE)
```

```python
import modelgen, find_architecture
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
lb.fit(labels_unique)
lb.classes_
```

```python
y_train_binary = lb.fit_transform(y_train)
y_val_binary = lb.fit_transform(y_val)
y_test_binary = lb.fit_transform(y_test)
```

```python
np.swapaxes(X_train,1,2).shape
```

```python
num_classes = y_train_binary.shape[1]

models = modelgen.generate_models(np.swapaxes(X_train,1,2).shape,
                                  number_of_classes=num_classes,
                                  number_of_models = 12)
```

```python
models
```

```python
models2 = [models[x] for x in range(1,12)]
len(models2)
```

```python
resultpath = ROOT
outputfile = os.path.join(resultpath, 'modelcomparison_Benchmarking1.json')
histories, val_accuracies, val_losses = find_architecture.train_models_on_samples(np.swapaxes(X_train,1,2), y_train_binary,
                                                                           np.swapaxes(X_val,1,2), y_val_binary,
                                                                           models2,nr_epochs=50,
                                                                           subset_size=-1,
                                                                           verbose=True,
                                                                           outputfile=outputfile)
print('Details of the training process were stored in ',outputfile)
```

```python
import pandas as pd

modelcomparisons = pd.DataFrame({'model':[str(params) for model, params, model_types in models2],
                                'model-type':[str(model_types) for model, params, model_types in models2],
                               'train_acc': [history.history['acc'][-1] for history in histories],
                               'train_loss': [history.history['loss'][-1] for history in histories],
                               'val_acc': [history.history['val_acc'][-1] for history in histories],
                               'val_loss': [history.history['val_loss'][-1] for history in histories]
                               })
modelcomparisons.to_csv(os.path.join(resultpath, 'modelcomparisons_Benchmarking1_01.csv'))

modelcomparisons
```

```python
ROOT
```

```python
PATH_models_trained = os.path.join(ROOT, 'models_trained')
filename_base = 'mcfly_model_'
model_types = [str(model_types) for model, params, model_types in models2]

for i, model in enumerate(models2):
    filename = filename_base + str(i) + "_" + model_types[i]
    filename = os.path.join(PATH_models_trained, filename)
    
    # serialize model to JSON
    model_json = model[0].to_json()
    with open(filename + ".json", "w") as json_file:
        json_file.write(model_json)
        
    # serialize weights to HDF5
    model[0].save(filename + ".hdf5")
    print("Saved model", i, "to disk")
```
